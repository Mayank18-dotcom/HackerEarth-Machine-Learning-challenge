```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('test.csv')
df1.head()
```

```python
df1.isnull().sum()
```

```python
med = df1['condition'].median()
med
```

```python
df1['condition'].fillna(med,inplace = True)
df1.isnull().sum()
```

```python
df1 = df1.drop('pet_id',axis = 'columns')
df1.head()
```

```python
df1['issue_date'] = pd.to_datetime(df1['issue_date'])
df1['listing_date'] = pd.to_datetime(df1['listing_date'])
df1.dtypes
```

```python
df1['time'] = df1['listing_date']-df1['issue_date']
df1.head()
```

```python
df1 = df1.drop(['issue_date','listing_date'],axis='columns')
df1.head(3)
```

```python
df1['time']=df1['time'].apply(lambda x: str(x))
df1.dtypes
```

```python
df1['time'] = df1['time'].apply(lambda x: int(x.split(' ')[0]))
df1.dtypes
```

```python
cname = df1['color_type'].unique()
cname
```

```python
df_dumm = pd.get_dummies(df1['color_type'],columns=cname)
df_dumm.head()
```

```python
df2 = df1.join(df_dumm)
df2.head()
```

```python
df2.drop('color_type',axis='columns')
df2.head(3)
```

```python
df2['height(cm)'] = df2['height(cm)'].apply(lambda x: x/100)
df2.head()
```

```python
df2.rename(columns = {'height(cm)':'height(m)'},inplace=True)
df2.head(3)
```

```python
df2.drop(['color_type'],axis = 'columns',inplace=True)
df2.head(3)
```

```python
from numpy.random import randint
df_empty1 = pd.DataFrame({'Black Tiger' : randint(1, size=8072),'Brown Tiger' : randint(1, size=8072)})
df_empty1
```

```python
df2 = df2.join(df_empty1)
df2.head()
```
